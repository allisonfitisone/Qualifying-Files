```python
def plot_3d_cone(A, color="purple"):
    C=Cone(A.rows())
    k = C.plot(wall_alpha=0.4, wall_color=color, ray_label=None, ray_thickness=1, wall_label=None, point_size=0)
    P = k + text3d("x", (2,0,0)) + text3d("y", (0,2,0)) + text3d("z", (0,0,2)) + line3d([(-2,0,0),(2,0,0)], color='black') + line3d([(0,-2,0),(0,2,0)], color='black') + line3d([(0,0,-2),(0,0,2)], color='black')
    return P
```

```python
def plot_3d_cone_decomposition(A, color1="blue", color2="yellow"):
    L=list(generate_cones_decomposition(A))
    print(L)
    c_1=Cone(list(L[0][0].rows()))
    c_2=Cone(list(L[1][0].rows()))
    g = c_2.plot(wall_alpha=0.4, wall_color=color1, ray_label=None, ray_thickness=1, wall_label=None, point_size=0) + c_1. plot(wall_alpha=0.4, wall_color=color2, ray_thickness=1, ray_label=None, wall_label=None, point_size=0)
    h=g+ text3d("x", (2,0,0)) + text3d("y", (0,2,0)) + text3d("z", (0,0,2)) + line3d([(-2,0,0),(2,0,0)], color='black') + line3d([(0,-2,0),(0,2,0)], color='black') + line3d([(0,0,-2),(0,0,2)], color='black')
    return h
```
