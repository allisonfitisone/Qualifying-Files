```python
# plot average time of solid angle measure (normal) wrt degree in purple
def plot_avg_time_sam_range(n, p, q,  A, eps=1e-6, simplicial=False, space="ambient"):
    logging.disable(logging.INFO)
    logging.disable(logging.WARNING)
    list = []
    for j in range(p,q+1):
        times = []
        for i in range(n):
            t=cputime()
            cputime(solid_angle_measure(A, eps=eps, deg=j, simplicial=simplicial, space=space))
            a_i = cputime(t)
            times.append(a_i)
        avg = sum(times) / n
        list_k = (j, avg)
        print(list_k)
        list.append(list_k)
    return list_plot(list, axes_labels=['$deg$', '$CPU time (s)$'], color=hue(0.8))
```

```python
# plot average time of solid angle measure (normal) wrt degree in orange
def plot_avg_time_sam_eps(n, A, s, t, deg=100, simplicial=False, space="ambient", color="red"):
    logging.disable(logging.INFO)
    logging.disable(logging.WARNING)
    list = []
    for j in range(s,t+1):
        e_j = 10**-j
        times = []
        for i in range(n):
            t=cputime()
            cputime(solid_angle_measure(A, eps=e_j, deg=deg, simplicial=simplicial, space=space))
            a_i = cputime(t)
            times.append(a_i)
        avg = sum(times) / n
        list_k = (j, avg)
        print(list_k)
        list.append(list_k)
    return list_plot(list, axes_labels=['$eps$', '$CPU time (s)$'], color=color)
```
